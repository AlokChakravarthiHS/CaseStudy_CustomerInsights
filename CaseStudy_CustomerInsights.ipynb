```python
# further by using this data we can make RFM model(Recency,Frequency,Monetary) Analysis
```

```python
#Recency grouping custID with presentdate and lastdate
latest_date=df_filtered['Date'].max()
print(latest_date)
```

```python
recency_df=df_filtered.groupby('Customer ID')['Date'].max().reset_index().rename(columns={'Date': 'LastDate'})
recency_df.head()
```

```python
recency_df['Recency']=recency_df['LastDate'].apply(lambda x:(latest_date - x).days)
recency_df.head(10)
```

```python
# Dropping 'LastDate' column 
recency_df.drop('LastDate',axis=1,inplace=True)
recency_df.head()
```

```python
#Frequency grouping custId and transactions
frequency_df=df_filtered.groupby('Customer ID')['Invoice'].count().reset_index()
frequency_df.head(10)
```

```python
#Monetary is grouping custid with TotalSpending
```

```python
# Group the data by customer ID total amount spent by each customer
monetary_df=df_filtered.groupby('Customer ID')['TotalSpending'].sum().reset_index()
monetary_df.head(10)
```

```python
rfm_df=pd.merge(recency_df,frequency_df,on='Customer ID')
rfm_df.head()
```

```python
rfm_df=pd.merge(rfm_df,monetary_df,on='Customer ID')
rfm_df.head()
```

```python
# Calc quantiles for the RFM 
quantiles=rfm_df[['Recency','Invoice','TotalSpending']].quantile(q=[0.25, 0.5, 0.75])
quantiles
```

```python
# converting to dict
quantiles_dict=quantiles.transpose().to_dict()
quantiles_dict
```

```python
# Create Scores of column by dividing Frequency into quartiles and assigning respective quantile labels

rfm_df['RecencyScore']=pd.qcut(rfm_df['Recency'],q=[0, 0.25, 0.5, 0.75, 1],labels= list(range(4, 0, -1)))
```

```python
# Create Scores of column by dividing Frequency into quartiles and assigning respective quantile labels

rfm_df['RecencyScore']=pd.qcut(rfm_df['Recency'],q = [0, 0.25, 0.5, 0.75, 1],labels = list(range(4, 0, -1)))
rfm_df['InvoiceScore']=pd.qcut(rfm_df['Invoice'],q=[0, 0.25, 0.5, 0.75, 1],labels=list(range(1, 5)))
rfm_df['TotalSpendingScore']=pd.qcut(rfm_df['TotalSpending'],q=[0, 0.25, 0.5, 0.75, 1],labels=list(range(1, 5)))
rfm_df.head()
```

```python
# To calc RFM score concat all the scores
rfm_df['RFMScore']=rfm_df['RecencyScore'].astype(str)+rfm_df['InvoiceScore'].astype(str)+rfm_df['TotalSpendingScore'].astype(str)
rfm_df.head()
```

```python
print('Best Customer:',len(rfm_df[rfm_df['RFMScore']=='444']))
print('Loyal Customers:',len(rfm_df[rfm_df['InvoiceScore']==4]))
print("Big Spenders:",len(rfm_df[rfm_df['TotalSpendingScore' ]==4]))
print('Almost Lost:', len(rfm_df[rfm_df['RFMScore']=='244']))
print('Lost Customers:',len(rfm_df[rfm_df['RFMScore']=='144']))
print('Lost Cheap Customers:',len(rfm_df[rfm_df['RFMScore']=='111']))
```
